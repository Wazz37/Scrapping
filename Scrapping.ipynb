```python
Generic_Name = ""
Salt = ""
Manufacturer = ""
Pack_size = ""
Price = ""
Benefits=""
Substitute = ""
Substitute_price = ""

data = [[Generic_Name,Salt,Manufacturer,Pack_size,Price,Benefits,Substitute]]
```

```python
for i in range(0,29):
    r = requests.get(url[i])
    soup = bs(r.content)
    contents = soup.prettify()

    medicine_details = soup.find(class_='medicine_details')
    med_rows = medicine_details.find_all("li")
   
    

    Generic_name = soup.find(class_='medicine_details')
    for index,first in enumerate(med_rows):
        if index == 1:
            Generic_Name = first.find("b").get_text()
    
    for index,row in enumerate(med_rows):
        if index == 1:
            Salt = row.find("b").get_text()
        elif index ==0:
            Manufacturer = row.find("a").get_text()
        else:
            exit
    size = soup.find(class_='middle_txt')

    for row1 in size:
        Pack_size = row1.replace("Infusion in 1 Bottle","ml")
        Pack_size = row1.replace("Liquid in 1 Bottle","")
        Pack_size = row1.replace("Injection in 1 Packet","")
        Pack_size = row1.replace("Gel in 1 Tube","")
        
    
    price = soup.find(class_="txt_big")
    Price=price.get_text()
    
    benefits = soup.find(class_='product_item_list')
    for row2 in benefits:
        Benefits=row2.get_text()
        
    substitute = soup.find_all(class_='product_item_list')
    for index,row3 in enumerate(substitute):
        if index==1:
            Substitute = row3.find("li").get_text()
    

    data.insert(i,[Generic_Name,Salt,Manufacturer,Pack_size,Price,Benefits,Substitute])
df = pd.DataFrame(data, columns = ['Generic Name','Salt','Manufacturer','Pack_size','Price','Benefits','Side_effects'])
```
